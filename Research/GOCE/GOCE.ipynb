# GOCE Gravity Data

Working with Colin on GOCE data.

```python
import geosoft.gxpy.gx as gx
import geosoft.gxpy.utility as gxu
import geosoft.gxpy.grid as gxgrd
import geosoft.gxpy.grid_utility as gxgrdu

from IPython.display import Image

gxc = gx.GXpy()
gxu.check_version('9.4.0b0')

grid = 'zz.grd'
grd = gxgrd.Grid.open(grid)
Image(grd.image_file(shade=True, pix_width=1000))
```

## TILT_ANGLE

See https://www.sciencedirect.com/science/article/pii/0926985194900221

```python
gdt = gxgrdu.derivative(grd, gxgrdu.TILT_ANGLE)
Image(gdt.image_file(shade=True, pix_width=1000))
```
